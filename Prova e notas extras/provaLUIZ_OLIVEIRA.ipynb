## Questão 1
Gere uma Lista contendo os primeiros 43 números pares.
Calcule sua soma.


```python
pares=[]
cont=1
soma=0
for i in range(0,100):
    if(cont<44):
        if(i%2==0):#condicao para par 
            pares.append(i)
            cont=cont+1
            soma=soma+i
    else:
        break
print(len(pares))
print(pares)
print(soma)
```
