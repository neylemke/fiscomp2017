## Questão 1
Gere uma Lista contendo os primeiros 32 números impares.
Calcule seu produto.


```python
lista=[]
for i in range(64):
    x=i
    if x%2!=0:
        lista.append(i)
print ('Lista= ', lista)

j=1
produto=[]
while j<len(lista): #orientação da monitora sobre interpretação do exercício
    p=lista[j]*lista[j-1]
    produto.append(p)
    j=j+1

print ('Produtos= ', produto)
```

## Questão 3
Gere um gráfico do peso e altura dos jogadores. Os eixos devem conter 
as quantidades que estão sendo representados e o título deve conter o nome do Aluno. 
O tamanho dos pontos deve representar a idade dos jogadores. A cor deve ser verde e
as marcas devem ser círculos.


```python
def read_jog_data(filename = "jogadores.csv"):
    import pandas as pd
    import numpy as np

    ## Read the .csv file with the pandas read_csv method
    jogs = pd.read_csv(filename)
    
    return jogs
jogs = read_jog_data()
```

```python
jogs.head()
```

```python
jogs.describe()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10,10)) # define plot area
ax = fig.gca() # define axis    
jogs['Peso'].plot.hist(ax = ax) # Use the plot.hist method on subset of the data frame
ax.set_title('Histograma dos jogadores') # Give the plot a main title
ax.set_xlabel('Peso') # Set text for the x axis
ax.set_ylabel('Numero de jogs')# Set text for y axis
```

```python
import seaborn as sns
sns.set_style("whitegrid")
sns.kdeplot(jogs['Peso'])
```

```python
fig = plt.figure(figsize=(10,10)) # define plot area
ax = fig.gca() # define axis 
sns.set_style("whitegrid")
sns.kdeplot(jogs['Peso'], ax = ax)
ax.set_title('Jogadores') # Give the plot a main title
ax.set_xlabel('Peso') # Set text for the x axis
ax.set_ylabel('Density')# Set text for y axis
```

```python
fig = plt.figure(figsize=(10,10)) # define plot area
ax = fig.gca() # define axis 
sns.set_style("whitegrid")
sns.violinplot(x = 'Peso', y = 'Altura', data = jogs, ax = ax)
ax.set_title('Jogadores') # Give the plot a main title
ax.set_xlabel('Peso') # Set text for the x axis
ax.set_ylabel('Altura')# Set text for y axis
```
