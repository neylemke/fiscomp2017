```python
# exercicio 1 
vet=[]
aux=0
i=0
while aux!=(46):
    if(i%2!=0):
        vet.append(i)
        aux+=1
    i+=1
print(vet)
print(len(vet))
b = 1
for i in range (len(vet)):
    b = b * vet[i]
print("O produto dos 46 números ímpares é: " +str(b))
```
