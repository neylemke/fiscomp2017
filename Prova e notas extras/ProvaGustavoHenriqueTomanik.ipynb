```python
#Exercicio1Gere uma Lista contendo os primeiros 48 números pares. Calcule seu produto.
lista = [ i  for i in range(2, 98, 2)]
print("Os valores da lista são: "+str(lista))
prod = 1
for i in range(len(lista)):
    prod = prod*lista[i]
print("O produto dos valores é: "+str(prod))
```
