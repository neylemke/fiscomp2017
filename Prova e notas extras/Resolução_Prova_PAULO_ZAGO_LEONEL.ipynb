```python
#Exercicio 1: Gere uma Lista contendo os primeiros 46 números impares. Calcule sua soma.

lista=[]
a=0
i=0
while a!=(46):
    if(i%2!=0):
        lista.append(i)
        a+=1
    i+=1
print("A lista com os primeiros 46 numeros impares é =", lista)

print("O tamanho da lista é =", len(lista))

for i in lista:
    a = a + i
print("A soma dos 46 primeiros numeros impares é =", a)
```
