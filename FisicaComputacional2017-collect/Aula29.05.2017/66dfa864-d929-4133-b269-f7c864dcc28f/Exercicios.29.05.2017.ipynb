```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
file=open("dracula.txt", "r")
```

```python
texto=file.readlines()
```

```python
dict={}
for line in texto:
    for word in line.split():
        if word in dict:
            dict[word]=dict[word]+1
        else:
            dict[word]=1
```

```python
oco=dict.values()
oco.sort()
X=np.array(range(len(oco)))+1
Y=np.array(oco)
```

```python
fig = plt.figure(figsize=(10,10)) # define plot area
ax = fig.gca() # define axis
x=fig.gca()
plt.plot(X,Y)
ax.set_title('Frequencia das palavras do Dracula') # Give the plot a main title
ax.set_xlabel('rank') # Set text for the x axis
ax.set_ylabel('freq')# Set text for y axis
```

```python
logx=np.log(X)
logy=np.log(Y)
fig = plt.figure(figsize=(10,10)) # define plot area
ax = fig.gca() # define axis
x=fig.gca()
plt.plot(logx,logy)
ax.set_title('Frequencia das palavras das Lusiadas') # Give the plot a main title
ax.set_xlabel('rank') # Set text for the x axis
ax.set_ylabel('freq')# Set text for y axis
```

```python
def read_auto_data(fileName = "Automobile price data.csv"):
    'Function to load the auto price data set from a .csv file' 
    import pandas as pd
    import numpy as np

    ## Read the .csv file with the pandas read_csv method
    auto_prices = pd.read_csv(fileName)
    
    ## Remove rows with missing values, accounting for mising values coded as '?'
    cols = ['price', 'bore', 'stroke', 
          'horsepower', 'peak-rpm']
    for column in cols:
        auto_prices.loc[auto_prices[column] == '?', column] = np.nan
    auto_prices.dropna(axis = 0, inplace = True)

    ## Convert some columns to numeric values
    for column in cols:
        auto_prices[column] = pd.to_numeric(auto_prices[column])
#    auto_prices[cols] = auto_prices[cols].as_type(int64)
        
    return auto_prices
auto_prices = read_auto_data()

import matplotlib.pyplot as plt
fig = plt.figure(figsize=(8, 8))
ax = fig.gca()
auto_prices.plot(kind = 'scatter', x = 'engine-size', y = 'price', ax = ax)
ax.set_title('Scatter plot de preco vs consumo')
ax.set_xlabel('Consumo') # Set text for the x axis
ax.set_ylabel('Preco')# Set text for y axis
```
