```python
file=open("new.txt", "r")
```

```python
texto=file.readlines()
```

```python
dict={}
for line in texto:
    for word in line.split():
        if word in dict:
            dict[word]=dict[word]+1
        else:
            dict[word]=1
```

```python
dict
```

```python
import numpy as np
oco=dict.values()
oco.sort()
oco.reverse()
x=np.array(range(len(oco)))+1
y=np.array(oco)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10,10))
ax = fig.gca()
x = fig.gca()
plt.plot(oco)
ax.set_title('Repeticao das palavras')
ax.set_xlabel('Ranking')
ax.set_ylabel('Repeticao')
```

```python
X=np.array(range(len(oco)))+1
Y=np.array(oco)
logx=np.log(X)
logy=np.log(Y)
fig = plt.figure(figsize=(10,10))
ax = fig.gca()
x = fig.gca()
ax.set_title('Repeticao das palavras (LOG)')
ax.set_xlabel('Ranking')
ax.set_ylabel('Repeticao')
plt.plot(logx,logy)
```

```python
def read_auto_data(fileName = "Automobile price data.csv"):
    'Function to load the auto price data set from a .csv file' 
    import pandas as pd
    import numpy as np

    ## Read the .csv file with the pandas read_csv method
    auto_prices = pd.read_csv(fileName)

    ## Remove rows with missing values, accounting for mising values coded as '?'
    cols = ['price', 'bore', 'stroke', 
          'horsepower', 'peak-rpm']
    for column in cols:
        auto_prices.loc[auto_prices[column] == '?', column] = np.nan
    auto_prices.dropna(axis = 0, inplace = True)

    ## Convert some columns to numeric values
    for column in cols:
        auto_prices[column] = pd.to_numeric(auto_prices[column])
#    auto_prices[cols] = auto_prices[cols].as_type(int64)

    return auto_prices
auto_prices = read_auto_data()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10, 10)) # define plot area
ax = fig.gca() # define axis
auto_prices.plot(kind = 'scatter', x = 'engine-size', y = 'price', ax = ax)
ax.set_title('Scatter plot de tamanho do moto vs preco') # Give the plot a main title
ax.set_xlabel('tamanho do motor') # Set text for the x axis
ax.set_ylabel('Preco (US)')# Set text for y axis
```
