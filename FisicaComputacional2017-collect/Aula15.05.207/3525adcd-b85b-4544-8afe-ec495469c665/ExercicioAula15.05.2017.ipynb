```python
#Considere uma matriz aleatória de 10×2 que represente coordenadas cartesianas, converta-as em coordenadas polares

import numpy as np
from numpy import linalg as LA
import math
from math import atan2
m = np.random.rand(10,2)*10
m = m.astype(int)
print("Coordenadas cartesianas:\n")
print(m)
print("\n")


x = list(m[:,0])
y = list(m[:,1])
print("Coordenadas polares:\n")
for i in range(len(m)):
    ro = (math.sqrt((pow(x[i], 2)) + (pow(y[i],2))))
    teta = math.atan2(y[i], x[i])
    xc = ro*math.cos(teta)
    yc = ro*math.sin(teta)
    print(xc, yc)
    
    




```

```python
#Considere um vetor aleatório com forma (100,2) representando coordenadas, encontre as distâncias entre todos os pontos
import numpy as np
from numpy import linalg as LA
v = np.random.rand(100,2)*10
v = v.astype(int)
for i in range(len(v)):
    for j in range(1, len(v)):
        d = LA.norm(v[i]-v[j])
        print("Distancia entre"+str(v[i])+"e"+str(v[j])+"é igual a= "+str(d))
```

```python
#Considere o vetor [1, 2, 3, 4, 5], como construir um novo vetor com 3 zeros consecutivos intercalados entre cada valor?

import numpy as np


v = np.array([1, 2, 3, 4, 5])
lista=[]

for i in range(len(v)):
    a=0
    lista.append(v[i])
    while a<3:
        lista.append(0)
        a = a+1
print(lista)

```
