```python
#Crie 12 arquivos, chamados de 'arq_mes_estação.txt'. 
#Cada um deles contendo a frase: A minha estação predileta é 'estação', especialmente o 'mês' repetida um número aleatório de vezes,
#entre 5 e 20.

from random import randint
mes=["janeiro",'fevereiro','marco','abril','maio','junho','julho','agosto','setembro','outubro','novembro','dezembro']
estacao=['verao','outono','inverno','primavera']
for i in range(12):
    arq=open("arq_"+str(mes[i])+"_"+str(estacao[randint(0,3)])+".txt","w")
    arq.write(randint(5,25)*"A minha estação predileta é "+str(estacao[randint(0,3)])+", especialmente o"+str(mes[i])+".\n")
    arq.close()
    !zip mes_estacao.zip arq*.txt
```
