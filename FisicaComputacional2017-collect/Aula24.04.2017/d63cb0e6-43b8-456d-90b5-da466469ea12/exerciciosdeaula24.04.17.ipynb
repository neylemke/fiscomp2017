```python
# Exercicio 1. Crie 12 arquivos, chamados 'arq_mes_estação.txt'. Cada um deles contendo a frase: A minha estação predileta é 'estação', especialmente 'mês' repetida de um numero aleatorio entre 5 e 20

m=['janeiro', 'fevereiro', 'março', 'abril', 'maio', 'junho', 'julho', 'agosto', 'setembro', 'outubro', 'novembro', 'dezembro']
e=['outono', 'inverno', 'primavera', 'verão']

from random import randrange
for i in range (1,13):
    if i>3 and i<=6:
        j=1
    elif i>=6 and i<=9:
            j=2
    elif i>=9 and i<=12:
            j=3
    else:
            j=4
                
    content= "Aminha estação predileta é"+str(e[j-1])+ "mês' repedida de um numero aleatorio de vezes entre 5 e 20.\n"
    with open('arq_'+str(m[i-1])+'_'+str(e[j-1])+'txt', 'w') as f:
         for k in range(1,randrange(6,20)):
            f.write(content)
    f.close()
    
    !zip file.zip arq*txt
```
