```python
import random

lista = []
for i in range(100):
    lista.append(random.randint(0,9))
print(lista)
for i in range(len(lista)-1):
    print ("Intervalo medio entre o numero "+str(lista[i])+" e "+str(lista[1+i])+" = "+str(lista[i+1] - lista[i]))
```
